Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Simular lanzamientos
num_lanzamientos = 100
lanzamientos = np.random.randint(1, 7, num_lanzamientos)

# Contar ocurrencias de cada número
counter = np.bincount(lanzamientos)

# Calcular las probabilidades aproximadas para 2, 3 y 5
# El index correspondiente es el número que quiero menos uno
probabilidades_2 = counter[1] / num_lanzamientos
probabilidades_3 = counter[2] / num_lanzamientos
probabilidades_5 = counter[4] / num_lanzamientos

print("Probabilidades 2:", probabilidades_2)
print("Probabilidades 3:", probabilidades_3)
print("Probabilidades 5:", probabilidades_5)

# Graficar el histograma de las probabilidades

# Ajustar nombres y probabilidades para que coincidan
nombres = ['0', '1', '2', '3', '4', '5', '6']
probabilidades = counter / num_lanzamientos

plt.bar(nombres, probabilidades)
plt.xlabel('Número en el dado')
plt.ylabel('Probabilidad')
plt.title('Histograma de probabilidades del dado equilibrado de 6 caras')

# Ajustar valores del eje y para mostrar porcentajes
plt.ylim(0, 1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.show()
```

5. Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio



**Ayuda Como generar los datos**


```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])  # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos


# Solicitar al usuario la cantidad N de datos a cargar
N = 1000

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)

# Crear una tabla con los datos
tabla_datos = np.array([
    ["Sexo", "Edad", "Altura", "Peso", "Horas Descanso", "Horas Trabajo/Estudio", "Horas Ejercicio"],
    *datos
])

# Imprimir tabla formateada
for row in tabla_datos:
    print("{:<15} {:<10} {:<10} {:<10} {:<15} {:<22} {:<15}".format(*row))


# Obtener columnas correspondientes a altura, peso, horas de descanso, trabajo y ejercicio
alturas = datos[:, 2].astype(float)  # Convert to float
pesos = datos[:, 3].astype(float)    # Convert to float
horas_descanso = datos[:, 4].astype(float)   # Convert to float
horas_trabajo_estudio = datos[:, 5].astype(float)  # Convert to float
horas_ejercicio = datos[:, 6].astype(float)        # Convert to float


# Calcular medias
media_altura = np.mean(alturas)
media_peso = np.mean(pesos)
media_horas_descanso = np.mean(horas_descanso)
media_horas_trabajo_estudio = np.mean(horas_trabajo_estudio)
media_horas_ejercicio = np.mean(horas_ejercicio)

print(f"Media de Altura: {media_altura:.2f}")
print(f"Media de Peso: {media_peso:.2f}")
print(f"Media de Horas de Descanso: {media_horas_descanso:.2f}")
print(f"Media de Horas de Trabajo/Estudio: {media_horas_trabajo_estudio:.2f}")
print(f"Media de Horas de Ejercicio: {media_horas_ejercicio:.2f}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(pesos, bins=20, color='blue', alpha=0.7)
plt.title('Distribución de Pesos')
plt.xlabel('Peso')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(alturas, bins=20, color='green', alpha=0.7)
plt.title('Distribución de Alturas')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 6))
plt.scatter(alturas, pesos, alpha=0.5)
plt.title('Correlación entre Altura y Peso')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.grid()
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(horas_descanso, label='Descanso', linewidth=2)
plt.plot(horas_trabajo_estudio, label='Trabajo/Estudio', linewidth=2)
plt.plot(horas_ejercicio, label='Ejercicio', linewidth=2)
plt.title('Horas de Descanso, Trabajo/Estudio y Ejercicio Diario')
plt.xlabel('Persona')
plt.ylabel('Horas')
plt.legend()
plt.grid()
plt.show()
```

```python
# Detectaremos y eliminaremos los outliers

def detect_outliers(data, column):
  q1 = data[column].quantile(0.25)
  q3 = data[column].quantile(0.75)
  iqr = q3-q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  return outliers

test_columns = ["Malic.acid","Ash","Acl","Mg","Proanth","Color.int", "Hue"]
outl_list = []
for c in test_columns:
  outliers = detect_outliers(df,c)
  outl_list.append(outliers)

# guardamos los outliers en un data frame

all_outliers = pd.concat(outl_list)

# rearmamos el data frame con los outliers eliminados

fixed_dataframe = df.drop(all_outliers.index)

# imprimimos cantidad de outliers:

print ("Encontramos " , len(all_outliers) ," outliers")

# imprimimos data frame nuevamente

fixed_dataframe.boxplot()
```
